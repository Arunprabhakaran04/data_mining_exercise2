```python
import pandas as pd
import numpy as np
```

```python
df_market = pd.read_csv("Market_Basket_Optimisation.csv")
df_market.head()
```

```python
df_market.shape
```

```python
df_market.info()
```

```python
df_market.fillna("",inplace=True)
df_market.isna().sum()
```

```python
df_market.head(10)
```

```python
df_market = df_market.map(lambda x: None if pd.isna(x) or x == "" else x)

transactions = df_market.apply(lambda row: [item for item in row.dropna()], axis=1)

transactions = transactions[transactions.apply(len) > 0]

print(transactions.head())
```

```python
from mlxtend.preprocessing import TransactionEncoder

# Convert transaction list to one-hot encoding format
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)

# Create a DataFrame from the encoded array
basket = pd.DataFrame(te_ary, columns=te.columns_)

basket.head()
```

```python
from mlxtend.frequent_patterns import fpgrowth

# Find frequent itemsets with minimum support of 2%
frequent_itemsets = fpgrowth(basket, min_support=0.02, use_colnames=True)

# Display frequent itemsets
frequent_itemsets.head(10)
```

```python
from mlxtend.frequent_patterns import association_rules

# Generate association rules using lift as the metric
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

# Display association rules
rules.head()
```

```python
rules.shape
```

```python
# Filter for strong association rules
strong_rules = rules[(rules['confidence'] >= 0.6) & (rules['lift'] >= 1.5)]

# Display the strongest rules
strong_rules.head()
```
