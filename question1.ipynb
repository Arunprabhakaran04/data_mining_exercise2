```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("Online_Retail.xlsx")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df = df.dropna(subset=['Description'])
df = df[df['Quantity'] > 0]
df = df.drop_duplicates()
print(df.info())
```

```python
# Choose the appropriate method
df = df.dropna(subset=['CustomerID'])  # Drop missing values (most common)
# OR df['CustomerID'] = df['CustomerID'].fillna(-1)  # Fill with placeholder

# Verify missing values
print(df.isnull().sum())  # Ensure no null values remain
```

```python
basket = df.groupby(['InvoiceNo', 'Description'])['Quantity'].sum().unstack().fillna(0)
basket = basket.applymap(lambda x: 1 if x > 0 else 0)
basket.head()
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
frequent_itemsets = apriori(basket, min_support=0.02, use_colnames=True)

# Generate association rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
```

```python
rules.head()
```

```python
rules.info()
```

```python
rules_confidence = rules.sort_values(by="confidence", ascending=False)
rules_confidence.head(10)
```

```python
rules_support = rules.sort_values(by="support", ascending=False)
rules_support.head(10)
```

```python
rules_lift = rules.sort_values(by="support", ascending=False)
rules_lift.head(10)
```

```python
#strong rules analysis - 
strong_rules = rules[(rules['confidence'] > 0.6) & (rules['lift'] > 1.2)]
```

```python
strong_rules.head(10)
```

```python
frequent_itemsets = fpgrowth(basket, min_support=0.02, use_colnames=True)

rules_fpgrowth = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
```
